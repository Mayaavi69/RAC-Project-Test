```python
# prompt: treat outliers where you are getting outliers in above mentioned columns

import numpy as np
# Select numeric columns except specified ones
numeric_cols = test.select_dtypes(include=np.number).columns.difference(['player_id', 'weak_foot', 'skill_moves'])

# Handle outliers for each numeric column
for col in numeric_cols:
    # Calculate IQR
    Q1 = test[col].quantile(0.25)
    Q3 = test[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier boundaries
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Replace outliers with median
    test[col] = np.where((test[col] < lower_bound) | (test[col] > upper_bound),
                         test[col].median(), test[col])

# Create boxplots for each numeric column after outlier treatment
for col in numeric_cols:
  fig = px.box(test, y=col, title=f"Boxplot of {col} (After Outlier Treatment)")
  fig.show()
```

```python
# prompt: split the complete data into testing(20%) and training(80%)

from sklearn.model_selection import train_test_split

# Assuming 'test' is your DataFrame and you want to split it into training and testing sets

# Split the data into 80% training and 20% testing
train_data, test_data = train_test_split(test, test_size=0.2, random_state=42)

# Now you have 'train_data' (80% of the data) and 'test_data' (20% of the data)
```

```python
# prompt: implement decision tree learner algorithm to the training dataset so that the machine learns and only do it on non categorical data except the player id column

import numpy as np
from sklearn.tree import DecisionTreeClassifier

# Select non-categorical columns except 'player_id' for training
features = train_data.select_dtypes(include=np.number).columns.difference(['player_id', 'cluster'])
X_train = train_data[features]
y_train = train_data['cluster']  # Assuming 'cluster' is the target variable

# Create and fit the Decision Tree model
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
# prompt: now implement the decision tree predictor to  the test data set

# Select the same features used during training
X_test = test_data[features]  # Use 'features' defined during training

# Make predictions on the test data
y_pred = clf.predict(X_test)

# Add the predictions to the test DataFrame
test_data['predicted_cluster'] = y_pred
```

```python
# prompt: now give me the visual of the decision tree

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Assuming 'clf' is your trained DecisionTreeClassifier

plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names=X_train.columns, class_names=[str(i) for i in clf.classes_])
plt.show()
```

```python
# prompt: give me the most top 5 important variable according to the descision tree

#import pandas as pd
# Get feature importances
importances = clf.feature_importances_

# Create a DataFrame to store feature importances
feature_importances_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': importances})

# Sort the DataFrame by importance in descending order
feature_importances_df = feature_importances_df.sort_values('Importance', ascending=False)

# Print the top 5 most important features
print(feature_importances_df.head(5))
```

```python
# prompt: now give me the confusion matrix and all its parameters

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report

# Assuming 'y_test' contains the true cluster labels for the test data
y_test = test_data['cluster']

# Calculate confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)

# Print classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
